## Sam conversion

For all the alignment output sam files:
- convert to bam, sort and filter
- Summarize the sizes for all the sam files / bam files / sorted bam files /
sorted and filtered bam files into a txt file named “sum_sizes.txt”

```python
%%bash
cd ~/ABABhwk/Unit2-module2-ChIPseq/sample_fastq/sample_fastq_trimmed/

for i in `ls *.sam | sed 's/.sam//g' | sort -u`
do
samtools view -bS $i.sam > $i.bam
samtools sort $i.bam -o $i.srt.bam
samtools rmdup -S $i.srt.bam $i.srt.dupr.bam
done
```

```python
%%bash
cd ~/ABABhwk/Unit2-module2-ChIPseq/sample_fastq/sample_fastq_trimmed/
ls -lh *.{sam,bam} | awk 'BEGIN{FS=" "}{print $5, $9}' >> sum_sizes.txt
cat sum_sizes.txt
```
