### MACS2

```python
%%bash
cd ~/ABABhwk/Unit2-module2-ChIPseq/hw4.3/

# MACS2 Peak calling
for i in *.srt.bam
do
macs2 callpeak -t $i -f BAM -n ${i/.bam/} -g mm -p 1e-2 --broad --nomodel --bdg --shift 0 --extsize 147 --keep-dup all &> ${i/.bam/.MACS2.out}
done
```

Difference between knownResults and homerResults:
- __knownResults__: knownResults are output of checking enrichment of known motifs. HOMER screens it's library of reliable motifs against the target and background sequences for enrichment, returning motifs enriched with a p-value less than 0.05. It is faster. Output files will have a certain sequence for each identified motif. It is particularly useful when the data is limited/signal is weak, since the knownResults will give more reliable results.
- __homerRsults__: homerResults are output of de novo motif search. By default, HOMER will search for motifs of len 8, 10, and 12 bp. It is slower. For each identified motif, it will have similar motif sequences. Comparing to knownResults, it allows you to directly query the sequence to discover which motifs are the most enriched sequences in the target set.
